```python
import os
```

```python
import matplotlib.pyplot as plt
import json
```

```python
runs_folder = '../WhenPigsFlyContext/v3_output/'
```

```python
os.listdir(runs_folder)
```

```python
set_1_experiments = ['lower_bound_material',
                '2_styles_hotswapped_material',
               '4_styles_hotswapped_material',
               '6_styles_hotswapped_material',
               '8_styles_hotswapped_material',
               '10_styles_hotswapped_material',
               'upper_bound_material']
```

```python
# sorted(os.listdir(exp_folder))
```

```python
dom_A_accuracies = []
dom_B_accuracies = []
for exp in set_1_experiments:
    exp_folder = os.path.join(runs_folder, exp)
    dom_A_file = os.path.join(exp_folder, 'test_domain_A_accuracies.json')
    dom_B_file = os.path.join(exp_folder, 'test_domain_B_accuracies.json')

    with open(dom_A_file, 'r') as F:
        dom_A_contents = json.load(F)
    
    with open(dom_B_file, 'r') as F:
        dom_B_contents = json.load(F)
    
    dom_A_accuracies.append(dom_A_contents['total_accuracy'])
    dom_B_accuracies.append(dom_B_contents['total_accuracy'])
```

```python
plt.figure(figsize=(10,6))
plt.plot(dom_A_accuracies)
plt.ylim(0,1)
plt.xticks(range(len(set_1_experiments)), labels=['lower bound', '2','4','6','8','10','upper bound'])
plt.xlabel('Number of styles')
plt.ylabel('Domain A Accuracy')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(dom_B_accuracies)
plt.ylim(0,1)
plt.xticks(range(len(set_1_experiments)), labels=['lower bound', '2','4','6','8','10','upper bound'])
plt.xlabel('Number of styles')
plt.ylabel('Domain B Accuracy')
plt.show()
```
