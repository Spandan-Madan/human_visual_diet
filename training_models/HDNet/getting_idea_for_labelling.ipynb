```python
import os
```

```python
crops_folder = '../DomainBed/domainbed/data/DGMaterials/main_xml_50/'
```

```python
categories = os.listdir(crops_folder)
```

```python
%matplotlib inline
```

```python
category_to_impaths = {}
for category in categories:
    category_folder = os.path.join(crops_folder, category)
    category_images = [os.path.join(category_folder,i) for i in os.listdir(category_folder)]
    category_to_impaths[category] = category_images
```

```python
categories
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import random
```

```python
cat = 'bed'
```

```python
sampled = random.sample(range(len(category_to_impaths[cat])), 10)

for sample in sampled:
    path = category_to_impaths[cat][sample]
    plt.imshow(Image.open(path))
    plt.show()
```
